```python
import numpy as np
# import matplotlib.pyplot as plt

from matplotlib import pyplot as plt

# matplotlib.pyplot.show() == pt.show()
```

```python
# اینجا یه تابع تعریف کردیم که دوتا ورودی رو میگیره و بر روی نمودار با استفاده از کتابخونه
# matplot
# نشون میده
def draw_function(x, y):
    plt.plot(x, y)
    plt.show()
```

$
\begin{equation}
  \text{Identity function} : f(x) = x
\end{equation}
$

```python
def identity_function(x):
    y = x
    return y
    
    
```

```python
# usage
x = [1, 3, 5, 10]

y = identity_function(x)
print(y)
```

```python
draw_function(x, y)
```

$
\begin{equation}
  \text{Quadratic function} : f(x) = ax^{2} + bx + c
\end{equation}
$

```python
def quadratic_function(x, a, b, c):
     
     #  دقیقا معادله بالا رو اومدیم نوشتیم
     y = a * (x **2) + b * x + c

     return y
```

```python
# usage
x = [ 1, 2, 3, 4, 5]

a, b, c = 1, 2, -1

y = quadratic_function(x, a, b, c)
```

```python
x = np.arange(-11, 11, 0.5)

# print(x)

a, b, c = -1, 2, -100

y = quadratic_function(x, a, b, c)

draw_function(x, y)
```

$
\begin{equation}
  \text{Sinusoidal function} : f(x) = sin(x)
\end{equation}
$

```python

def sinx(x, a = 1, b = 1, c = 0):
    y = a * np.sin(b * x + c)

    # y = np.sin(x)
    return y
```

```python
# usage
# x = np.arange(-100, 100, 0.5) 
x = [1, 2, 3, 4]
a, b, c = -1, 2, -100

y = sinx(x, a, b, c)

draw_function(x, y)
```
